# Color of Image

```python
img2 = plt.imread('../data/img/city1.jpg')
```

```python
plt.imshow(img2)
plt.show()
```

```python
plt.imshow(img2)
plt.show()

plt.figure(figsize=(10, 10))
plt.subplot(1,3,1)
plt.title('R')
plt.imshow(img2[:, :, 0], 'gray')

plt.subplot(1,3,2)
plt.title('G')
plt.imshow(img2[:, :, 1], 'gray')

plt.subplot(1,3,3)
plt.title('B')
plt.imshow(img2[:, :, 2], 'gray')
plt.show()
```
