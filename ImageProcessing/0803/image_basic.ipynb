# COLOR  

```python
path = '/Users/jeong-yeobin/VScode/bigdatacamp_image_processing/ImageProcessing/data/img/city1.jpg'

color_img = plt.imread(path)
plt.imshow(color_img)
plt.show()
```

```python
plt.title('color')
plt.imshow(color_img)
plt.show()

plt.figure(figsize=(10, 10))
plt.subplot(1,3,1)
plt.title('red')
plt.imshow(color_img[:, :, 0], 'gray')

plt.subplot(1,3,2)
plt.title('green')
plt.imshow(color_img[:, :, 1], 'gray')

plt.subplot(1,3,3)
plt.title('blue')
plt.imshow(color_img[:, :, 2], 'gray')

plt.show()
```
